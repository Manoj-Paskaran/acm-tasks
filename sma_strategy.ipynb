```python
import yfinance as yf
import numpy as np
import plotly.express as px
```

```python
df = yf.download("TCS.NS", period="6mo", multi_level_index=False)
```

```python
df["SMA_5"] = df["Close"].rolling(window=5).mean()
df["SMA_20"] = df["Close"].rolling(window=20).mean()

df['Signal'] = 0
# df['Signal'][df['SMA_5'] > df['SMA_20']] = 1
df.loc[df['SMA_5'] > df['SMA_20'], "Signal"] = 1 
# df['Signal'][df['SMA_5'] < df['SMA_20']] = -1

df.loc[df['SMA_5'] < df['SMA_20'], "Signal"] = -1 
df['Position'] = df['Signal'].diff()
df.iloc[30:60]
```

```python
# Add a column to mark Buy and Sell signals for Plotly
df["Buy_Signal"] = df["Position"] == 2
df["Sell_Signal"] = df["Position"] == -2

# Create the Plotly Express figure
fig = px.line(
    df,
    x=df.index,
    y=["Close", "SMA_5", "SMA_20"],
    labels={"value": "Price", "variable": "Legend"},
    title="TCS - SMA Crossover Strategy"
)

# Add Buy signals
fig.add_scatter(
    x=df[df["Buy_Signal"]].index,
    y=df["SMA_5"][df["Buy_Signal"]],
    mode="markers",
    marker=dict(symbol="triangle-up", size=10, color="green"),
    name="Buy Signal"
)

# Add Sell signals
fig.add_scatter(
    x=df[df["Sell_Signal"]].index,
    y=df["SMA_5"][df["Sell_Signal"]],
    mode="markers",
    marker=dict(symbol="triangle-down", size=10, color="red"),
    name="Sell Signal"
)

# Update layout for better visuals
fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Price",
    legend_title="Legend",
    template="plotly_white",
    height=600,
    width=1000
)
fig
```
